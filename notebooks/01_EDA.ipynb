# EDA — FER2013 Emotion Dataset

```python
import sys; sys.path.insert(0, '..')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from data.dataloader import auto_load, EMOTIONS
sns.set_style('darkgrid')
```

```python
train_ds, val_ds, test_ds, class_weights, (X_test, y_test) = auto_load('../data/raw', batch_size=64)
print('Class weights:', class_weights)
```

```python
# Load raw CSV for label counts
import os
csv_path = '../data/raw/fer2013.csv'
if os.path.exists(csv_path):
    df = pd.read_csv(csv_path)
    counts = df['emotion'].value_counts().sort_index()
    plt.figure(figsize=(10,4))
    bars = plt.bar(EMOTIONS, counts.values, color=sns.color_palette('Set2', 7))
    for bar, val in zip(bars, counts.values):
        plt.text(bar.get_x()+bar.get_width()/2, bar.get_height()+50, str(val), ha='center', fontsize=10)
    plt.title('Class Distribution — FER2013 (imbalanced!)', fontsize=14)
    plt.ylabel('Number of images'); plt.tight_layout(); plt.show()
```

```python
# Sample images per emotion
fig, axes = plt.subplots(7, 8, figsize=(16, 14))
for cls_idx, emotion in enumerate(EMOTIONS):
    idxs = np.where(y_test.argmax(1) == cls_idx)[0][:8]
    for j, idx in enumerate(idxs):
        axes[cls_idx, j].imshow(X_test[idx, ..., 0], cmap='gray')
        axes[cls_idx, j].axis('off')
        if j == 0:
            axes[cls_idx, j].set_ylabel(emotion, fontsize=10)
plt.suptitle('Sample Images per Emotion Class', fontsize=14)
plt.tight_layout(); plt.show()
```

```python
# Pixel intensity distribution
plt.figure(figsize=(10, 4))
for cls_idx, emotion in enumerate(EMOTIONS):
    idxs = np.where(y_test.argmax(1) == cls_idx)[0]
    pixels = X_test[idxs].flatten()
    sns.kdeplot(pixels, label=emotion)
plt.title('Pixel Intensity Distribution per Emotion')
plt.xlabel('Pixel value (normalized)'); plt.legend(); plt.tight_layout(); plt.show()
```

```python
# Mean face per emotion
fig, axes = plt.subplots(1, 7, figsize=(16, 3))
for cls_idx, (emotion, ax) in enumerate(zip(EMOTIONS, axes)):
    idxs = np.where(y_test.argmax(1) == cls_idx)[0]
    mean_face = X_test[idxs].mean(axis=0)[..., 0]
    ax.imshow(mean_face, cmap='gray')
    ax.set_title(emotion); ax.axis('off')
plt.suptitle('Mean Face per Emotion', fontsize=13)
plt.tight_layout(); plt.show()
```
