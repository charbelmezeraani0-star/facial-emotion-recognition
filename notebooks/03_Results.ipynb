# Results — Evaluation, Grad-CAM, Ensemble

```python
import sys; sys.path.insert(0, '..')
import tensorflow as tf
from data.dataloader import auto_load
from evaluation.evaluate import evaluate_model, plot_confusion_matrix, compare_models
from evaluation.gradcam import visualize_gradcam_all_classes
from models.ensemble import evaluate_ensemble

_, _, _, _, (X_test, y_test) = auto_load('../data/raw', batch_size=64)

cnn    = tf.keras.models.load_model('../saved_models/custom_cnn_best.keras')
mobile = tf.keras.models.load_model('../saved_models/mobilenetv2_finetune_best.keras')
resnet = tf.keras.models.load_model('../saved_models/resnet50_finetune_best.keras')
```

```python
results = {}
for name, model in [('CustomCNN', cnn), ('MobileNetV2', mobile), ('ResNet50', resnet)]:
    acc, f1, y_true, y_pred = evaluate_model(model, X_test, y_test, name)
    plot_confusion_matrix(y_true, y_pred, name)
    results[name] = (acc, f1)

# Ensemble
ens_acc, _ = evaluate_ensemble([cnn, mobile, resnet], X_test, y_test)
results['Ensemble'] = (ens_acc, None)
compare_models(results)
```
